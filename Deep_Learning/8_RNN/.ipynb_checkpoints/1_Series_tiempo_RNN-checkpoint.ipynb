# RNN in time series

When dealing with sequential or time series data, traditional feedforward networks cannot be used for learning and prediction. A mechanism to retain past or historical information is required to forecast future values. Recurrent neural networks, or RNNs for short, are a variant of conventional feedforward artificial neural networks that can handle sequential data and can be trained to maintain knowledge about the past.



```python
from pandas import read_csv
import numpy as np
from keras.models import Sequential
from keras.layers import Dense, SimpleRNN
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
import math
import matplotlib.pyplot as plt
```

## Keras SimpleRNN

The following function returns a model that includes a SimpleRNN layer and a dense layer to learn sequential data. The input_shape specifies the parameter (time_steps x features). We will simplify everything and use univariate data, that is, a single characteristic; the time steps are discussed below.

```python
def create_RNN(hidden_units, dense_units, input_shape, activation):
    model = Sequential()
    model.add(SimpleRNN(hidden_units, input_shape=input_shape, 
                        activation=activation[0]))
    model.add(Dense(units=dense_units, activation=activation[1]))
    model.compile(loss='mean_squared_error', optimizer='adam')
    return model

demo_model = create_RNN(2, 1, (3,1), activation=['linear', 'linear'])
demo_model.summary()
```

The demo_model object with two hidden units created via the SimpleRNN  and one dense unit created via the Dense layer. The input_shape is set to 3 Ã— 1, and a linear activation function is used on both layers for simplicity. Just to remind you, the linear activation function does not change the input.
Let's look at the above weights. Note: Since the weights are randomly initialized, the results here will be different than yours. The important thing is to learn what the structure of each object that is used looks like and how it interacts with others to produce the final result.


```python
wx = demo_model.get_weights()[0]
wh = demo_model.get_weights()[1]
bh = demo_model.get_weights()[2]
wy = demo_model.get_weights()[3]
by = demo_model.get_weights()[4]

print('wx = ', wx, ' wh = ', wh, ' bh = ', bh, ' wy =', wy, 'by = ', by)
```

Now let's do a simple experiment to see how the SimpleRNN and Dense layers produce an output. Keep this figure in sight.

<img src="imgs/rnnCode1.png" scale="150" align="left" > 

We will input $x$ for three time steps and let the network generate an output. The values of the hidden units at time steps 1, 2 and 3 will be calculated. The $h0$ vector is initialized to zero values. The output is calculated from $y$. An activation function is not required since we are using linear units

```python
x = np.array([1, 2, 3])
# Reshape the input to the required sample_size x time_steps x features 
x_input = np.reshape(x,(1, 3, 1))
#usar el modelo generado demo_model
y_pred_model = demo_model.predict(x_input)

#hacer el proceso paso a paso, usando la idea presentada en la figura anterior
m = 2
h0 = np.zeros(m)
h1 = np.dot(x[0], wx) + h0 + bh
h2 = np.dot(x[1], wx) + np.dot(h1,wh) + bh
h3 = np.dot(x[2], wx) + np.dot(h2,wh) + bh
o3 = np.dot(h3, wy) + by

print('h1 = ', h1,'h2 = ', h2,'h3 = ', h3)
#comparar las salidas
print("Prediction from network ", y_pred_model)
print("Prediction from our computation ", o3)
```

# RNN with sunspots database

Now that we understand how the SimpleRNN and Dense layers come together. Let's run a full RNN on a simple time series dataset. We will have to follow these steps:

     1. Read the dataset from a given URL
     2. Split the data into training and test sets
     3. Prepare input to the required Keras format
     4. Create an RNN model and train it
     5. Make the predictions on test and training sets and print the root mean square error on both sets
     6. View the result
    
**Background:** Sunspots are temporary phenomena in the Sun's photosphere that appear as spots darker than the surrounding areas. They are regions of reduced surface temperature caused by concentrations of magnetic field flux that inhibit convection. Sunspots usually appear in pairs of opposite magnetic polarity. Their number varies according to the solar cycle of approximately 11 years.


source: https://en.wikipedia.org/wiki/Sunspot

# Steps 1, 2: 

The following function reads training and test data from a csv file, then splits it into a given percentage of training and test data. Returns one-dimensional arrays for training and testing data after scaling the data between 0 and 1 using scikit-learn's MinMaxScaler.

```python
#
def get_train_test(csv_file, split_percent=0.8):
    df = read_csv(csv_file, usecols=[1], engine='python')
    data = np.array(df.values.astype('float32'))
    scaler = MinMaxScaler(feature_range=(0, 1))
    data = scaler.fit_transform(data).flatten()
    n = len(data)
    # Point for splitting data into train and test
    split = int(n*split_percent)
    train_data = data[range(split)]
    test_data = data[split:]
    return train_data, test_data, data

sunspots_url = 'sunspots/monthly-sunspots.csv'
train_data, test_data, data = get_train_test(sunspots_url)
```

# Step 3: 

The next step is to prepare the data for training the Keras model. The input matrix must have the following form:  total_samples x time_steps x characteristics.

There are many ways to prepare time series data for training. We will create input rows with non-overlapping time steps. An example is shown in the following figure for time steps = 2. Here, time steps denote the number of previous time steps that will be used to predict the next value of the time series data.

<img src="imgs/rnnCode2.png" scale="60" align="left" > 

The function get_XY()  takes a one-dimensional array as input and converts it to the required X input and Y response arrays. We will use 12 time_steps for the sunspot data set, since sunspots typically have a 12-month cycle. You can experiment with other values of time_steps.



```python
# Prepare the input X and target Y
def get_XY(dat, time_steps):
    # Indices of target array
    Y_ind = np.arange(time_steps, len(dat), time_steps)
    Y = dat[Y_ind]
    # Prepare X
    rows_x = len(Y)
    X = dat[range(time_steps*rows_x)]
    X = np.reshape(X, (rows_x, time_steps, 1))    
    return X, Y

time_steps = 12
trainX, trainY = get_XY(train_data, time_steps)
testX, testY = get_XY(test_data, time_steps)
```

## Step 4: 

For this step, you can reuse your create_RNN() function that was defined earlier.

```python
model = create_RNN(hidden_units=3, dense_units=1, input_shape=(time_steps,1), 
                   activation=['tanh', 'tanh'])
model.fit(trainX, trainY, epochs=20, batch_size=1, verbose=2)
```

## Step 5: 

The print_error() function calculates the root mean square error between the actual and predicted values.

```python
def print_error(trainY, testY, train_predict, test_predict):    
    # Error of predictions
    train_rmse = math.sqrt(mean_squared_error(trainY, train_predict))
    test_rmse = math.sqrt(mean_squared_error(testY, test_predict))
    # Print RMSE
    print('Train RMSE: %.3f RMSE' % (train_rmse))
    print('Test RMSE: %.3f RMSE' % (test_rmse))    

# make predictions
train_predict = model.predict(trainX)
test_predict = model.predict(testX)
# Mean square error
print_error(trainY, testY, train_predict, test_predict)
```

## step 6: 

The following function plots the actual target values and the predicted values. The red line separates the training and test data points.

```python
# Plot the result
def plot_result(trainY, testY, train_predict, test_predict):
    actual = np.append(trainY, testY)
    predictions = np.append(train_predict, test_predict)
    rows = len(actual)
    plt.figure(figsize=(15, 6), dpi=80)
    plt.plot(range(rows), actual)
    plt.plot(range(rows), predictions)
    plt.axvline(x=len(trainY), color='r')
    plt.legend(['Actual', 'Predictions'])
    plt.xlabel('Observation number after given time steps')
    plt.ylabel('Sunspots scaled')
    plt.title('Actual and Predicted Values. The Red Line Separates The Training And Test Examples')
plot_result(trainY, testY, train_predict, test_predict)
```
