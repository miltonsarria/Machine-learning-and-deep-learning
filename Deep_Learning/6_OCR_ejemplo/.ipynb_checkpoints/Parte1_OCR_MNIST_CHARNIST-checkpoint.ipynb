# OCR

El reconocimiento óptico de caracteres, conocido como OCR por sus siglas en inglés (Optical Character Recognition), es una tecnología que permite la conversión de documentos físicos o imágenes escaneadas en texto digital editable. Utilizando algoritmos avanzados, el OCR interpreta patrones visuales de caracteres y los transforma en datos electrónicos que pueden ser procesados por computadoras. Esta herramienta es fundamental en la digitalización de documentos, facilitando la búsqueda, edición y almacenamiento de información de manera eficiente y precisa. El OCR encuentra aplicaciones en una amplia gama de campos, desde la automatización de procesos administrativos hasta la preservación y accesibilidad de documentos históricos.

El conjunto de datos MNIST estándar está integrado en marcos de aprendizaje profundo populares, incluidos Keras, TensorFlow, PyTorch, etc. En la Figura 1 (izquierda) se puede ver una muestra del conjunto de datos MNIST 0-9. El conjunto de datos MNIST nos permitirá reconocer los dígitos del 0 al 9. Cada uno de estos dígitos está contenido en una imagen en escala de grises de 28 x 28. 

Pero ¿qué pasa con las letras A-Z? El conjunto de datos estándar MNIST no incluye ejemplos de los caracteres A-Z: ¿cómo vamos a reconocerlos?

La respuesta es utilizar la base de datos **especial NIST 19**, que incluye caracteres A-Z. En realidad, este conjunto de datos cubre 62 caracteres hexadecimales ASCII correspondientes a los dígitos del 0 al 9, las letras mayúsculas de la A a la Z y las letras minúsculas de la A a la Z.

Para que el conjunto de datos sea más fácil de usar, el usuario de Kaggle, Sachin Patel, lo ha publicado en un archivo CSV fácil de usar. Este conjunto de datos toma las letras mayúsculas A-Z de la base de datos especial 19 de NIST y las reescala para que sean 28 x 28 píxeles en escala de grises para que tengan el mismo formato que nuestros datos MNIST.

se puede descargar del siguiente enlace: https://www.kaggle.com/datasets/sachinpatel21/az-handwritten-alphabets-in-csv-format

Implementaremos métodos y utilidades que nos permitirán:

     1. Cargar ambos conjuntos de datos para MNIST 0-9 dígitos y letras Kaggle A-Z desde el disco
     2. Combinar estos conjuntos de datos en un conjunto de datos de caracteres unificado
     3. Manejar el sesgo/desequilibrio de la etiqueta de clase por tener un número diferente de muestras por carácter
     4. Entrenar un modelo de Keras y TensorFlow en el conjunto de datos combinado
     5. Mostrar los resultados del entrenamiento y visualizar el resultado de los datos de validación.

```python
### cargar librerias necesarias
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.optimizers import SGD
from sklearn.preprocessing import LabelBinarizer
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from imutils import build_montages
import matplotlib.pyplot as plt
import numpy as np
import argparse
import cv2
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Dropout
import matplotlib.pyplot       as plt
from mlxtend.data import loadlocal_mnist
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.utils import to_categorical
```

```python
#load data MNIST
root_dir = ''

X_train, y_train = loadlocal_mnist(
        images_path=root_dir+'MNIST/train-images-idx3-ubyte', 
        labels_path=root_dir+'MNIST/train-labels-idx1-ubyte')

X_test, y_test = loadlocal_mnist(
        images_path=root_dir+'MNIST/t10k-images-idx3-ubyte', 
        labels_path=root_dir+'MNIST/t10k-labels-idx1-ubyte')

digitsData = np.vstack([X_train, X_test])
digitsLabels     = np.hstack([y_train, y_test])
    
    
digitsData = digitsData.reshape(70000,28,28,1)
labels,count_class = np.unique(digitsLabels,return_counts=True)


print('[INFO] \nlabels: \n %s \ncount per class \n %s' % (labels,count_class))
print(f'Data set dimensions: {(digitsData.shape)}')   
    
```

```python
#funcion para cargar datos A-Z
def load_az_dataset(datasetPath):
	# initialize the list of data and labels
	data = []
	labels = []
	# loop over the rows of the A-Z handwritten digit dataset
	for row in open(datasetPath):
		# parse the label and image from the row
		row = row.split(",")
		label = int(row[0])
		image = np.array([int(x) for x in row[1:]], dtype="uint8")
		# images are represented as single channel (grayscale) images
		# that are 28x28=784 pixels -- we need to take this flattened
		# 784-d list of numbers and repshape them into a 28x28 matrix
		image = image.reshape((28, 28))
		# update the list of data and labels
		data.append(image)
		labels.append(label)
	# convert the data and labels to NumPy arrays
	data = np.array(data, dtype="float32")
	labels = np.array(labels, dtype="int")
	# return a 2-tuple of the A-Z data and labels
	return (data, labels)
```

```python
# cargar A-Z 
print("[INFO] loading datasets...")
(azData, azLabels) = load_az_dataset("A_Z Handwritten Data.csv")

azData = azData.reshape(372451,28,28,1)
labels,count_class = np.unique(azLabels,return_counts=True)

print('[INFO] \nlabels: \n %s \ncount per class \n %s' % (labels,count_class))
print(f'Data set dimensions: {(azData.shape)}')
```

A medida que combinamos nuestras letras y números en un conjunto unificado, queremos eliminar cualquier ambigüedad donde haya superposición en las etiquetas para que cada etiqueta en el conjunto de caracteres combinado sea única.

Actualmente, nuestras etiquetas de la A a la Z van desde [0, 25], correspondientes a cada letra del alfabeto. Las etiquetas de nuestros dígitos van del 0 al 9, por lo que hay superposición, lo que sería problemático si las combináramos directamente.

Simplemente agregaremos diez a todas nuestras etiquetas A-Z para que todas tengan valores de etiqueta enteros mayores que nuestros valores de etiqueta de dígitos. Ahora tenemos un esquema de etiquetado unificado para los dígitos del 0 al 9 y las letras de la A a la Z sin superposición en los valores de las etiquetas.

Si va a usar un modelo pre-entrenado, por ejemplo la arquitectura ResNet requiere que las imágenes tengan dimensiones de entrada de 32 x 32, pero nuestras imágenes de entrada actualmente tienen un tamaño de 28 x 28. Cambiamos el tamaño de cada una de las imágenes usando cv2.resize, solamente si es necesario



```python
# La base de datos MNIST tiene etiquetas de 0 a 9, se le suma +10  a las etiquetas de 
#las letras de a-z 
azLabels += 10
# Unir los datos en un solo conjunto
data = np.vstack([azData, digitsData])
labels = np.hstack([azLabels, digitsLabels])

# se cambia el tamanio a  32x32
#data = [cv2.resize(image, (32, 32)) for image in data]
data = np.array(data, dtype="float32")
#ajustar para que sea de 3 canales
#data = np.repeat(data[:, :,:, np.newaxis], 3, axis=3)

#data = np.expand_dims(data, axis=-1)#si es un solo canal y no se ha agregado 

data /= 255.0
print(f"data dimensions {data.shape}, labels{labels.shape}")
```

Nuestro siguiente paso es preparar las etiquetas para la CNN, ponderar las etiquetas para tener en cuenta el sesgo en el número de veces que cada clase (carácter) se representa en los datos y dividir los datos en conjuntos de prueba y entrenamiento:

```python
le = LabelBinarizer()
labels = le.fit_transform(labels)
counts = labels.sum(axis=0)
# account for skew in the labeled data
classTotals = labels.sum(axis=0)
classWeight = {}
# loop over all classes and calculate the class weight
for i in range(0, len(classTotals)):
    classWeight[i] = classTotals.max() / classTotals[i]
# partition the data into training and testing splits using 80% of
# the data for training and the remaining 20% for testing
(trainX, testX, trainY, testY) = train_test_split(data,
    labels, test_size=0.20, stratify=labels, random_state=42)

plt.figure(figsize=(10, 5))
images =  np.zeros((28,1))
for i in range(40):
    image = trainX[i]    
    images=np.append(images,image[:,:,0],axis=1)

plt.imshow(images,cmap="gray");
plt.tick_params(left = False, right = False , labelleft = False ,
                labelbottom = False, bottom = False) 
```

Podemos mejorar los resultados de nuestro clasificador aumentando los datos de entrada para el entrenamiento utilizando un ImageDataGenerator.

Se incluyen varias rotaciones, escalado de tamaño, traslaciones horizontales, traslaciones verticales e inclinaciones en las imágenes.

```python
aug = ImageDataGenerator(
                        rotation_range=10,
                        zoom_range=0.05,
                        width_shift_range=0.1,
                        height_shift_range=0.1,
                        shear_range=0.15,
                        horizontal_flip=False,
                        fill_mode="nearest")
EPOCHS = 50 #numero de epocas
INIT_LR = 1e-1 #tasa de aprendizaje inicial
BS = 128       #tamanio de lote
num_clases = len(le.classes_) #cuantas clases?
print("[INFO] compiling model...")

# cargar modelo ResNet50 
'''
base_model = ResNet50(weights='imagenet', include_top=False, input_shape=(32, 32, 3))
opt = SGD(learning_rate=INIT_LR, decay=INIT_LR / EPOCHS)
# Congelar capas
for layer in base_model.layers:
    layer.trainable = False

# Crear nuevo modelo (clasificador)
model = Sequential()
model.add(base_model)
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(num_clases, activation='softmax')) 
'''

# Define el modelo CNN
model = models.Sequential()

# Primera capa convolucional
model.add(layers.Conv2D(64, (3, 3), activation='relu', input_shape=(28, 28, 1)))
model.add(layers.MaxPooling2D((2, 2)))

# Segunda capa convolucional
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))

# Flatten Layer : para pasar los datos a la FC (FC - fully connected)
model.add(layers.Flatten())

# Fully Connected Layer
model.add(layers.Dense(128, activation='relu'))
model.add(Dropout(0.5))
# Capa de salida
model.add(layers.Dense(num_clases, activation='softmax'))  # En este caso se tienen 10 clases 

model.compile(loss="categorical_crossentropy", optimizer=opt,metrics=["accuracy"])
print("[INFO] done...")
```

```python
print("[INFO] training network...")
H = model.fit(
              aug.flow(trainX, trainY, batch_size=BS),
              validation_data=(testX, testY),
              steps_per_epoch=len(trainX) // BS,
              epochs=EPOCHS,
              class_weight=classWeight,
              verbose=1)
# define la lista de etiquetas a reconocer, recuerden que primero van los numeros luego letras a-z
labelNames = "0123456789"
labelNames += "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
#almacenarlos en una lista
labelNames = [l for l in labelNames]
# evaluate the network
print("[INFO] evaluating network...")
predictions = model.predict(testX, batch_size=BS)
print(classification_report(testY.argmax(axis=1),
                            predictions.argmax(axis=1), 
                            target_names=labelNames))
```

El historial de entrenamiento se puede ver a continuación:

```python

print("[INFO] serializing network...")
#guardar el modelo
model.save('model_HandWrite.h5', save_format="h5")

# mostrar hostorial
N = np.arange(0, EPOCHS)
plt.style.use("ggplot")
plt.figure()
plt.plot(N, H.history["loss"], label="train_loss")
plt.plot(N, H.history["val_loss"], label="val_loss")
plt.title("Training Loss and Accuracy")
plt.xlabel("Epoch #")
plt.ylabel("Loss/Accuracy")
plt.legend(loc="lower left")
```

A continuación se presentan algunos resultados de muestra de nuestro conjunto de pruebas:

```python
# inicializar una lista vacia
images = []
# seleccionar de forma aleatoria algunas muestras del conjunto de prueba
for i in np.random.choice(np.arange(0, len(testY)), size=(49,)):
    # clasificar la imagen
    probs = model.predict(testX[np.newaxis, i])
    prediction = probs.argmax(axis=1)
    label = labelNames[prediction[0]]
    # extraer la imagen e inicializar el color para poner la etiqueta correcta
    image = (testX[i] * 255).astype("uint8")
    color = (0, 255, 0)
    # Si hay error usar otro color
    if prediction[0] != np.argmax(testY[i]):
        color = (255, 0, 0)
    # aumentar el tamanio de la imagen de 28x28 a 50 x 50 para verla mejor, y repetir los canales
    # poner la prediccion de la imagen con texto
    image = cv2.merge([image] * 3)
    image = cv2.resize(image, (50, 50), interpolation=cv2.INTER_LINEAR)
    cv2.putText(image, label, (5, 20), cv2.FONT_HERSHEY_SIMPLEX, 0.75,color, 2)
    # agregar la imagen a la lista
    images.append(image)
# construir el montaje a mostrar
montage = build_montages(images, (50, 50), (7, 7))[0]
# mostrar el resultado
plt.figure(figsize=(8,8))
plt.imshow(montage)
```
