## Detección de colores en OpenCV

1. Leer el documento adjunto sobre espacios de color.

Paso 1: Imagen a procesar
Lo primero que necesitamos es una imagen o fotograma para sobre este detectar los colores. Podemos conseguir nuestra «materia prima» de dos formas, leyendo una imagen o a través de un video. 

Usaremos PIL para mostrar los resultados

Paso 2: Transformar de BGR a HSV

Por defecto OpenCV lee a las imágenes o fotogramas en BGR, por ello es necesario transformarlas al espacio de color HSV. Para ello nos ayudaremos de la función cv2.cvtColor
, como primer argumento le daremos la imagen a transformar, y luego para indicar que transformaremos de BGR a HSV.

```python
import cv2                               #Importamos el modulo opencv 
import numpy as np
from PIL import Image                    #Importamos el modulo PIL
import matplotlib.pyplot as plt
frame = cv2.imread('bd_frutas/cerezas/cherry-pair-fruits-sweet-162689.jpeg')

##usar pil para mostrar es necesario pasar de BGR a RGB
img = Image.fromarray(cv2.cvtColor(frame,cv2.COLOR_BGR2RGB))
img.show()
```

```python
#pasar de bgr a hsv
frame = cv2.imread('bd_frutas/cerezas/cherry-pair-fruits-sweet-162689.jpeg')
frameHSV = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)  


##usar pil para mostrar
img = Image.fromarray(frameHSV) #es necesario usar fromarray para convertir de matrices a un objeto PIL
img.show()
```

Paso 3: Determinar los rangos en donde se encuentra el color a detectar
<img src="img/gyuw4.png" scale="1">

Como vemos en la figura  el componente en H va de 0 a 179, y en este pasan colores de rojo, naranja, amarillo, verde, azul, violeta y nuevamente rojo, entonces vamos a determinar 2 rangos para el color rojo que está presente al principio y al final, el primer rango lo determinaremos de 0 a 8, y el segundo de 175 a 179 en H, mientras que para los componente S de 100 a 255, y V de 20 a 255, para ambos rangos.

**Pregunta** Que implicaciones tiene que se puedaseleccionar S y V en los rangos descritos?


se usa la función cv2.inRange, el primer argumento es la imagen en la cual se buscará el rango, en este caso frameHSV, luego el límite inicial y final del primer rango, el mismo proceso se realizará  para el segundo rango. 
De aquí se obtendrán las imágenes maskRed1 y maskRed2, que son binarias en donde el color blanco refleja el lugar en donde está presente el primer y segundo rango respectivamente, mientras que el color negro mostrará el lugar donde no están presentes estos rangos.

Como deseamos detectar el color rojo, y tenemos 2 rangos establecidos, debemos unirlos. maskRed por lo tanto también será binaria.

```python
#definir rangos
redBajo1 = np.array([0, 100, 20], np.uint8)
redAlto1 = np.array([8, 255, 255], np.uint8)

redBajo2=np.array([175, 100, 20], np.uint8)
redAlto2=np.array([179, 255, 255], np.uint8)

#pasar de bgr a hsv 
frame = cv2.imread('bd_frutas/fresas/strawberry-7224875_1280.jpg')
frameHSV = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)  
#crear dos mascaras una para el rango alto del rojo otra para el rango bajo
maskRed1 = cv2.inRange(frameHSV, redBajo1, redAlto1)
maskRed2 = cv2.inRange(frameHSV, redBajo2, redAlto2)
# Aquí estamos sumando ambas mascaras para mostrar en una sola (maskRed), la presencia del color rojo
maskRed = cv2.add(maskRed1, maskRed2)


##mascara  como es binaria no es necesario convertir de bgr a rgb
img=Image.fromarray(maskRed)
img.show()
```

Paso 4: Visualización

Entonces tenemos una imagen binaria, en donde el color blanco representa el color rojo detectado según los rangos que hemos dado en el paso 3. Es hora de visualizar este resultado.

```python
maskRedvis = cv2.bitwise_and(frame, frame, mask= maskRed)   


##usar pil para mostrar maskRedvis, recordar que se debe pasar de bgr a rgb
img = Image.fromarray(cv2.cvtColor(maskRedvis,cv2.COLOR_BGR2RGB))
img.show()
```

La detección de colores puede ayudarte a realizar un montón de aplicaciones, sin embargo hay dos aspectos importantes a considerar:

   - Iluminación. Es de importancia tener controlado este aspecto, ya que a más o menos iluminación sobre el objeto o la región que deseamos detectar cierto color, puede alterar el resultado de dicha detección.
   - Fondo. El fondo de la imagen también podría irrumpir en el proceso de detección, ya que si en este están presentes colores que deseamos detectar, no solo se detectaría el objeto de interés, sino también parte del fondo.

## Detección de rostro y ojos

Cuando trabajamos en un problema de clasificación facial, podríamos aplicar detección de rostros para validar la imagen (¿aparece alguna cara con un mínimo de resolución?), recortarla (para aislar la parte interesante) y por ejemplo enderezarla

Primero elegimos los clasificadores pre-entrenados en la detección de rostro y de ojos. Hay una lista de archivos XML disponibles que podemos usar. Hay que tener en cuenta que no hay uno de ellos que sea el mejor en todos los casos :

1) Para la detección de rostro, OpenCV ofrece los siguientes (desde el a priori más relajado hasta el más estricto):

    haarcascade_frontalface_default.xml
    haarcascade_frontalface_alt.xml
    haarcascade_frontalface_alt2.xml
    haarcascade_frontalface_alt_tree.xml

2) Para la detección de ojos, podemos elegir entre dos:

    haarcascade_eye.xml
    haarcascade_eye_tree_eyeglasses.xml (¡trata de lidiar con las gafas!)



**IMPORTANTE!!: En la actualidad existen otros sistemas más efectivos basados en redes neuronales, los que se usan en esta guia son métodos clásicos** 


Nosotros cargamos los clasificadores pre-entrenados de esta manera usando una de las opciones tanto para rostro como para ojos:


```python
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
eyes_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_eye.xml')
```

```python
#Para la detección de rostro usamos este sencillo código:
faces_detected = face_cascade.detectMultiScale(frame, scaleFactor=1.1, minNeighbors=5)
```

El resultado es una lista que contiene todas las caras detectadas, o mejor dicho la posición de los rectángulos que las delimitan. Podemos visualizarlo fácilmente:

**Pregunta** Cómo funciona rectangle de opencv? dar algunos ejemplos...

```python
(x, y, w, h) = faces_detected[0] 
cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 1); 


##usar pil
img = Image.fromarray(frame)
img.show()
```

Para los ojos procedemos de una manera similar, pero reduciendo la búsqueda al rectángulo de la cara:

```python
eyes = eyes_cascade.detectMultiScale(frame[y:y+h, x:x+w])
for (ex, ey, ew, eh) in eyes:
    cv2.rectangle(frame, (x+ex, y+ey), (x+ex+ew, y+ey+eh),  (255, 255, 255), 1)

##usar pil
img = Image.fromarray(frame)
img.show()    
```

## Transformaciones geométricas
podemos recortar y rotar una imagen

```python
#definir el tamaño
lx, ly = im2.size
#definir una caja (esquina superior izquierda, esquina superior derecha)
box1 = (lx//4,ly//4,lx-lx//4,ly-ly//4) 
#hacer el recorte
crop = im2.crop(box1)
#mostrar la imagen de abajo a arriba con flipud
plt.subplot(121)
flip_crop= np.flipud(crop)
plt.imshow(flip_crop, cmap=plt.cm.gray)

## girar 45 grados
plt.subplot(122)
rotate_crop = ndimage.rotate(crop, 45)
plt.imshow(rotate_crop, cmap=plt.cm.gray)
```

## Filtrado de imágenes

Filtros locales: reemplaza el valor de los pixeles por una función de los valores de los pixeles vecinos.

Vecindad: cuadrado (selecciona el tamaño), círculo o elementos estructurados más complejos*.

2. enfoque/refinamiento (sharpening): Enfocar una imagen borrosa

primero hacemos una imagen borrosa y luego tratamos de refinarla

**ejercicio** variar el parámetro alpha. Que efecto observa??

```python
blurred_l = ndimage.gaussian_filter(im2, sigma=3)

#Incrementamos el peso de los bordes agregando una aproximación del laplaciano
filter_blurred_l = ndimage.gaussian_filter(blurred_l, sigma=3)
alpha = 2
sharpened = blurred_l + alpha * (blurred_l - filter_blurred_l)

plt.figure(figsize=(20,10))
##
plt.subplot(131)
plt.imshow(im2, cmap=plt.cm.gray)

## 
plt.subplot(132)
plt.imshow(blurred_l, cmap=plt.cm.gray)

## 
plt.subplot(133)
plt.imshow(sharpened, cmap=plt.cm.gray)
```

## Reduciendo ruido
primero vamos a agregarle ruido a la imagen. Primero recortamos una zona de la imagen, luego agregamos ruido aleatorio y visualizamos las dos imagenes para comparar

```python
## recortar la zona del ojo izquierdo 
box1 = (500,100,600,250) 
#hacer el recorte
zona_ojo = im2.crop(box1)

##creamos una imagen ruidosa
img = np.asarray(zona_ojo)  #la covertimos a array de np
noisy = img + 0.9*img.std()*np.random.random(img.shape)

plt.figure(figsize=(15,5))
##
plt.subplot(121)
plt.imshow(img, cmap=plt.cm.gray)

## 
plt.subplot(122)
plt.imshow(noisy, cmap=plt.cm.gray)
```

Un filtro gaussiano suaviza la imagen eliminando el ruido... además de los bordes. 
La mayoría de los filtros lineales isotrópicos locales (ndimage.uniform_filter) desenfocan la imagen.
Un filtro de mediana conserva mejor los bordes


```python
gauss_denoised = ndimage.gaussian_filter(noisy, 2)
med_denoised = ndimage.median_filter(noisy, 5)

plt.figure(figsize=(15,5))
##
plt.subplot(131)
plt.imshow(noisy, cmap=plt.cm.gray)
plt.title("imagen ruidosa")
## 
plt.subplot(132)
plt.imshow(gauss_denoised, cmap=plt.cm.gray)
plt.title("Filrado gausiano")
## 
plt.subplot(133)
plt.imshow(med_denoised, cmap=plt.cm.gray)
plt.title("Filrado de mediana")
```

## Diferenciar patrones

A continuación se presentan los pasos tipicos para preparar una base de datos que servira posteriormente para entrenar un clasificador. Usaremos dos tipos de imagenes, flechas a la derecha y flechas a la izquierda. 

1. Se cargan las imágenes
2. Se pasan a escala de gris
3. Todas las imagenes tienen diferente tamaño, entonces se convierten a un mismo tamaño de 30 x 30
4. Se aplica una técnica llamada zca (zero component analysis) para normalizar las imagenes, previamente se divide entre 255 para garantizar amplitud de 0 a 1

```python
import numpy as np
from glob import glob
import os
from pathlib import Path
from PIL import Image, ImageOps 
from process import *

carpeta = "flechas/"
file_names = glob(carpeta+"**/*.png",recursive=True)
clases     = np.array(["izquierda","derecha"])
frames    = []
labels    =[]

new_dim = (30,30) #dar las nuevas dimensiones (alto, ancho)


for file_name in file_names:
    frame = Image.open(file_name)
    frame = ImageOps.grayscale(frame)
    frame = frame.resize(new_dim)
    frame = np.asarray(frame).ravel()
    
    clase=Path(file_name).parts[1]
    label = np.where(clases==clase)[0][0]
    labels.append(label)
    frames.append(frame)

frames=np.array(frames)/255

normalizer = zca()
normalizer.fit(frames)

new_f = normalizer.transform(frames)   

img = new_f[1].reshape((30,30))
plt.imshow(img, cmap=plt.cm.gray)
```

# Ver los histogramas resultantes

```python
plt.figure(figsize=(20,5))


for i,img in enumerate(new_f):
    if labels[i]==0:
        plt.subplot(1,7,i+1)
        b=plt.hist(img,100)

plt.figure(figsize=(20,5))


for i,img in enumerate(new_f):
    if labels[i]==1:
        plt.subplot(1,7,(i+1)-7)
        b=plt.hist(img,100)
```

# Ver las imagenes  resultantes ya normalizadas

```python
plt.figure(figsize=(20,5))


for i,img in enumerate(new_f):
    if labels[i]==0:
        plt.subplot(1,7,i+1)
        img = img.reshape((30,30))
        plt.imshow(img, cmap=plt.cm.gray)

plt.figure(figsize=(20,5))


for i,img in enumerate(new_f):
    if labels[i]==1:
        plt.subplot(1,7,(i+1)-7)
        img = img.reshape((30,30))
        plt.imshow(img, cmap=plt.cm.gray)
  
```
